**4-NETTOYAGE DES DONNÉES :**

**4-1-Supprimer les valeurs incorrectes**

```python
print("=== ÉTAPE 4-1 : Suppression des valeurs incorrectes ===\n")

# 1. Vérifier les valeurs aberrantes dans les colonnes numériques
colonnes_numeriques = df.select_dtypes(include=['int64', 'float64']).columns

print("Valeurs aberrantes détectées :")
for col in colonnes_numeriques:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if len(outliers) > 0:
        print(f"  - {col} : {len(outliers)} valeurs aberrantes")
        # Remplacer les outliers par la médiane
        mediane = df[col].median()
        df.loc[df[col] < lower_bound, col] = mediane
        df.loc[df[col] > upper_bound, col] = mediane
        print(f"    Remplacées par la médiane ({mediane})")

# 2. Vérifier les valeurs incorrectes dans les colonnes catégorielles
colonnes_categoriques = df.select_dtypes(include=['object']).columns

print("\nValeurs incorrectes dans les colonnes catégorielles :")
for col in colonnes_categoriques:
    valeurs_uniques = df[col].unique()
    print(f"  - {col} : {len(valeurs_uniques)} valeurs uniques")
    # Afficher les valeurs uniques pour vérification
    if len(valeurs_uniques) <= 20:
        print(f"    {valeurs_uniques.tolist()}")

print(f"\n Dataset après suppression des valeurs incorrectes : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

**4-2-Convertir les types de données :**

```python
print("\n=== ÉTAPE 4-2 : Conversion des types de données ===\n")

print("Types avant conversion :")
print(df.dtypes)

# Convertir la date
if 'application_date' in df.columns:
    df['application_date'] = pd.to_datetime(df['application_date'])
    print("\n application_date convertie en datetime")

# Convertir les booléens
if 'consent_flag' in df.columns:
    df['consent_flag'] = df['consent_flag'].astype(bool)
    print(" consent_flag convertie en bool")

# Convertir les colonnes catégorielles en 'category'
colonnes_categoriques = df.select_dtypes(include=['object']).columns
for col in colonnes_categoriques:
    if col != 'credit_risk':
        df[col] = df[col].astype('category')
        print(f" {col} convertie en category")

print("\nTypes après conversion :")
print(df.dtypes)
```

**4-3 Encoder les variables catégorielles :**

```python
print("\n=== ÉTAPE 4-3 : Encodage des variables catégorielles ===\n")

# Créer une copie du dataset pour l'encodage
df_encoded = df.copy()

# 1. Encoder la variable cible 'credit_risk'
print("1. Encodage de la variable cible 'credit_risk' :")
le_target = LabelEncoder()
df_encoded['credit_risk_encoded'] = le_target.fit_transform(df_encoded['credit_risk'])
print(f"   - good → {le_target.transform(['good'])[0]}")
print(f"   - bad → {le_target.transform(['bad'])[0]}")
print(f"    credit_risk encodée")

# 2. Encoder les variables catégorielles binaires
print("\n2. Encodage des variables binaires :")
colonnes_binaires = ['gender', 'marital_status', 'housing_status', 'consent_flag']
for col in colonnes_binaires:
    if col in df_encoded.columns:
        le = LabelEncoder()
        df_encoded[col + '_encoded'] = le.fit_transform(df_encoded[col])
        print(f"    {col} encodée (valeurs: {dict(zip(le.classes_, le.transform(le.classes_)))})")

# 3. One-Hot Encoding pour les variables catégorielles non binaires
print("\n3. One-Hot Encoding pour les variables multicatégories :")
colonnes_onehot = ['region', 'education_level', 'employment_status', 'loan_purpose']
colonnes_onehot = [col for col in colonnes_onehot if col in df_encoded.columns]

df_encoded = pd.get_dummies(df_encoded, columns=colonnes_onehot, prefix=colonnes_onehot, drop_first=True)
print(f"    {len(colonnes_onehot)} colonnes encodées en One-Hot")

print(f"\n Dataset après encodage : {df_encoded.shape[0]} lignes, {df_encoded.shape[1]} colonnes")
```

**5-PROTÉGER LES DONNÉES PERSONNELLES :**

**5-1-Supprimer les identifiants directs :**

```python
print("\n=== ÉTAPE 5-1 : Suppression des identifiants directs ===\n")

# Liste des identifiants directs à supprimer
identifiants_directs = ['customer_id', 'full_name', 'email', 'phone', 'national_id']

# Vérifier quelles colonnes existent
colonnes_a_supprimer = [col for col in identifiants_directs if col in df_encoded.columns]
print(f"Colonnes à supprimer : {colonnes_a_supprimer}")

# Supprimer les colonnes
df_encoded = df_encoded.drop(columns=colonnes_a_supprimer)
print(f" {len(colonnes_a_supprimer)} colonnes supprimées")

print(f"Dataset après suppression : {df_encoded.shape[0]} lignes, {df_encoded.shape[1]} colonnes")
```

**5-2-Anonymiser les données financières**

```python
print("\n=== ÉTAPE 5-2 : Anonymisation des données financières ===\n")

# Colonnes financières à anonymiser
colonnes_financieres = ['annual_income', 'savings_balance', 'checking_balance', 'loan_amount']

print("Anonymisation par normalisation (StandardScaler) :")
scaler = StandardScaler()

for col in colonnes_financieres:
    if col in df_encoded.columns:
        # Sauvegarder les valeurs originales pour référence (optionnel)
        df_encoded[col + '_original'] = df_encoded[col]

        # Normaliser les valeurs
        df_encoded[col + '_normalized'] = scaler.fit_transform(df_encoded[[col]])
        print(f"   {col} normalisée (moyenne=0, écart-type=1)")

        # Supprimer la colonne originale
        df_encoded = df_encoded.drop(columns=[col])

print(f"\n Dataset après anonymisation : {df_encoded.shape[0]} lignes, {df_encoded.shape[1]} colonnes")
```

**5-3-Pseudonymiser les quasi-identifiants :**

```python
print("\n=== ÉTAPE 5-3 : Pseudonymisation des quasi-identifiants ===\n")
# 1. Agréger l'âge en tranches
if 'age' in df_encoded.columns:
    print("1. Agrégation de l'âge en tranches :")
    bins = [0, 25, 35, 45, 55, 100]
    labels = ['18-25', '26-35', '36-45', '46-55', '56+']
    df_encoded['age_group'] = pd.cut(df_encoded['age'], bins=bins, labels=labels, right=False)
    df_encoded = df_encoded.drop(columns=['age'])
    print(f"   Âge agréé en {len(labels)} tranches : {labels}")
    print(f"   Distribution :\n{df_encoded['age_group'].value_counts().sort_index()}")

# 2. Catégoriser les revenus (si non déjà normalisés)
print("\n2. Catégorisation des revenus :")
if 'annual_income_normalized' in df_encoded.columns:
    # Créer des catégories de revenus basées sur les percentiles
    df_encoded['income_category'] = pd.qcut(df_encoded['annual_income_normalized'],
                                             q=4,
                                             labels=['Très bas', 'Bas', 'Élevé', 'Très élevé'])
    print(f"   Revenus catégorisés en 4 classes")
    print(f"   Distribution :\n{df_encoded['income_category'].value_counts().sort_index()}")

print(f"\n Dataset après pseudonymisation : {df_encoded.shape[0]} lignes, {df_encoded.shape[1]} colonnes")
```

**5-4-Masquage des données sensibles :**

```python
print("\n=== ÉTAPE 5-4 : Masquage des données sensibles ===\n")

# Créer un identifiant générique pour les clients
if 'customer_id' in df.columns:
    print("1. Création d'un identifiant générique :")
    df_encoded['customer_reference'] = ['USER' + str(i).zfill(4) for i in range(1, len(df_encoded) + 1)]
    print(f"  Customer ID masqué : {df_encoded['customer_reference'].iloc[:5].tolist()} ...")

# Masquer les noms (si présents)
if 'full_name' in df.columns:
    print("\n 2. Masquage des noms :")
    df_encoded['full_name_masked'] = ['Client_' + str(i) for i in range(1, len(df_encoded) + 1)]
    print(f"  Noms masqués : {df_encoded['full_name_masked'].iloc[:5].tolist()} ...")

# Masquer les emails (si présents)
if 'email' in df.columns:
    print("\n 3. Masquage des emails :")
    df_encoded['email_masked'] = ['email_' + str(i) + '@masked.com' for i in range(1, len(df_encoded) + 1)]
    print(f"   Emails masqués : {df_encoded['email_masked'].iloc[:5].tolist()} ...")

print(f"\n Dataset après masquage : {df_encoded.shape[0]} lignes, {df_encoded.shape[1]} colonnes")
```

```python
# ÉTAPE : CRÉER LE DATASET ML-READY

print("\n" + "="*60)
print("ÉTAPE 15 : CRÉATION DU DATASET ML-READY")
print("="*60)

print("\n 15.1 : Préparation du dataset final")

# Définir les colonnes utiles pour le modèle
colonnes_utiles = [
    # Quasi-identifiants pseudonymisés
    'age_group',
    'gender_encoded',
    'marital_status_encoded',
    'housing_status_encoded',
    'income_category',
    'customer_reference',

    # Données financières anonymisées
    'annual_income_normalized',
    'savings_balance_normalized',
    'checking_balance_normalized',
    'loan_amount_normalized',
    'loan_duration_months',
    'existing_credits',
    'past_due_count',
    'credit_history_score',
    'debt_ratio',

    # One-Hot encodées
    'loan_purpose_business',
    'loan_purpose_car',
    'loan_purpose_education',
    'loan_purpose_health',
    'loan_purpose_home_improvement',
    'loan_purpose_housing',
    'loan_purpose_other',

    # Variable cible
    'credit_risk_encoded'
]

# Ajouter les colonnes One-Hot existantes pour region, education, employment
for col in df_encoded.columns:
    if col.startswith(('region_', 'education_level_', 'employment_status_')):
        if col not in colonnes_utiles:
            colonnes_utiles.append(col)

# Garder uniquement les colonnes qui existent
colonnes_existantes = [col for col in colonnes_utiles if col in df_encoded.columns]
df_ml = df_encoded[colonnes_existantes]

print(f" Dataset ML-ready : {df_ml.shape[0]} lignes, {df_ml.shape[1]} colonnes")

# 15.2 : Vérification finale
print("\n 15.2 : Vérification finale")

print("  1. Statistiques descriptives :")
print(f"      - Moyenne d'âge : {df_ml['age_group'].value_counts().index[0] if 'age_group' in df_ml.columns else 'N/A'}")
print(f"      - Revenu moyen (normalisé) : {df_ml['annual_income_normalized'].mean():.2f}" if 'annual_income_normalized' in df_ml.columns else "")
print(f"      - Dette moyenne : {df_ml['debt_ratio'].mean():.2f}" if 'debt_ratio' in df_ml.columns else "")

print("\n  2. Distribution de la variable cible :")
if 'credit_risk_encoded' in df_ml.columns:
    print(f"      - 0 (good) : {(df_ml['credit_risk_encoded'] == 0).sum()} clients")
    print(f"      - 1 (bad) : {(df_ml['credit_risk_encoded'] == 1).sum()} clients")

print("\n  3. Valeurs manquantes :")
print(f"      - {df_ml.isnull().sum().sum()} valeurs manquantes totales")

print("\n  4. Aperçu du dataset :")
display(df_ml.head())

# 15.3 : Sauvegarder le dataset ML-ready
print("\n 15.3 : Sauvegarde du dataset ML-ready")

# Créer le dossier data/processed
os.makedirs('data/processed', exist_ok=True)

# Sauvegarder localement dans Colab
df_ml.to_csv('data/processed/credit_processed.csv', index=False)
print("   Dataset sauvegardé localement : data/processed/credit_processed.csv")


# TÉLÉCHARGER VERS L'ORDINATEUR LOCAL

print("\n" + "="*60)
print(" TÉLÉCHARGEMENT VERS VOTRE ORDINATEUR")
print("="*60)

print("\nChoisissez l'emplacement sur votre ordinateur pour sauvegarder le fichier :")

# Option 1 : Télécharger directement via le navigateur
print("\n OPTION 1 : Télécharger via le navigateur")
print("   Cliquez sur le bouton ci-dessous pour télécharger le fichier")
files.download('data/processed/credit_processed.csv')

print("\n Fichier téléchargé sur votre ordinateur !")

# Option 2 : Sauvegarder avec un nom personnalisé
print("\n" + "="*60)
print(" SAUVEGARDE AVEC NOM PERSONNALISÉ")
print("="*60)

print("\nVoulez-vous renommer le fichier avant de le sauvegarder ?")
print("Par défaut : credit_processed.csv")

# Proposer de renommer
custom_name = input("\nEntrez un nouveau nom (ou appuyez sur Entrée pour garder le nom par défaut) : ")

if custom_name.strip():
    custom_filename = custom_name.strip()
    if not custom_filename.endswith('.csv'):
        custom_filename += '.csv'
else:
    custom_filename = 'credit_processed.csv'

# Sauvegarder avec le nouveau nom
df_ml.to_csv(custom_filename, index=False)
print(f" Dataset sauvegardé sous : {custom_filename}")

# Télécharger avec le nouveau nom
print(f"\n Téléchargement de {custom_filename}...")
files.download(custom_filename)
print(f" {custom_filename} téléchargé sur votre ordinateur !")
```

```python
"""
ÉTAPES : CALCUL DES HASH SHA-256 (Version simplifiée)
"""
import hashlib
import os
from google.colab import files
from datetime import datetime

print("="*60)
print("CALCUL DES HASH SHA-256")
print("="*60)

# Créer le dossier hashes
os.makedirs('hashes', exist_ok=True)

# Fonction pour calculer le hash
def calculate_sha256(file_path):
    sha256_hash = hashlib.sha256()
    with open(file_path, "rb") as f:
        for byte_block in iter(lambda: f.read(4096), b""):
            sha256_hash.update(byte_block)
    return sha256_hash.hexdigest()

# ÉTAPE 19 : Hash du dataset brut

print("\n ÉTAPE 19 : Hash du dataset brut")
print(" Upload du fichier brut (credit.csv ou data_credit_risk.csv) :")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Calculer le hash
hash_brut = calculate_sha256(filename)

# Sauvegarder
with open('hashes/raw_dataset_hash.txt', 'w') as f:
    f.write("HASH SHA-256 - DATASET BRUT\n")
    f.write("="*50 + "\n")
    f.write(f"Fichier : {filename}\n")
    f.write(f"Date : {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write(f"Hash : {hash_brut}\n")

print(f"\n Hash brut : {hash_brut}")
print(f" Sauvegardé : hashes/raw_dataset_hash.txt")
files.download('hashes/raw_dataset_hash.txt')

# ÉTAPE 20 : Hash du dataset traité

print("\n" + "="*60)
print("ÉTAPE 20 : Hash du dataset traité")

processed_file = 'data/processed/credit_processed.csv'

if os.path.exists(processed_file):
    hash_traite = calculate_sha256(processed_file)

    with open('hashes/processed_dataset_hash.txt', 'w') as f:
        f.write("HASH SHA-256 - DATASET TRAITÉ\n")
        f.write("="*50 + "\n")
        f.write(f"Fichier : {processed_file}\n")
        f.write(f"Date : {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Hash : {hash_traite}\n")

    print(f"\n Hash traité : {hash_traite}")
    print(f"Sauvegardé : hashes/processed_dataset_hash.txt")
    files.download('hashes/processed_dataset_hash.txt')
else:
    print("\n Dataset traité non trouvé !")
    print("Upload du fichier traité (credit_processed.csv) :")
    uploaded_processed = files.upload()
    processed_filename = list(uploaded_processed.keys())[0]

    hash_traite = calculate_sha256(processed_filename)

    with open('hashes/processed_dataset_hash.txt', 'w') as f:
        f.write("HASH SHA-256 - DATASET TRAITÉ\n")
        f.write("="*50 + "\n")
        f.write(f"Fichier : {processed_filename}\n")
        f.write(f"Date : {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Hash : {hash_traite}\n")

    print(f"\n Hash traité : {hash_traite}")
    print(f"Sauvegardé : hashes/processed_dataset_hash.txt")
    files.download('hashes/processed_dataset_hash.txt')

print("\n" + "="*60)
print("TOUS LES HASH ONT ÉTÉ TÉLÉCHARGÉS !")
print("="*60)
```
